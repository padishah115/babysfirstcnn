```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(123)
#Defines a convolutional layer which takes 3 input channels (3) and moves to a feature map with 16 dimensions
conv = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)

#Convolutional kernel matrix is randomised at startup
conv.weight
```

```python
torch.manual_seed(123)
input_tensor = torch.rand((3, 32, 32)) #RGB image with 32x32 pixel layers

input_tensor.shape
```

```python
output_tensor = conv(input_tensor)

#output_tensor, 
output_tensor.shape
```

```python
output_tensor = F.max_pool2d(F.max_pool2d(input=input_tensor, kernel_size=2), kernel_size=2)

output_tensor.shape
```
